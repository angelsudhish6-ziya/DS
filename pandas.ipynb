```python
import pandas as pd

df1 = pd.DataFrame({
    'key': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie']
})

df2 = pd.DataFrame({
    'key': [1, 2, 3],
    'Age': [25, 30, 35]
})

result = pd.merge(df1, df2, on='key')
print(result)
```
